```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data_caffein = pd.read_csv('/caffeine 2.csv')

sns.set_theme()
sns.set(rc={"figure.dpi" : 300})
sns.set(rc={"figure.figsize" : (6,3)})

sns.scatterplot(x="Volume (ml)", y="Calories", data = data_caffein, hue='type')
```
